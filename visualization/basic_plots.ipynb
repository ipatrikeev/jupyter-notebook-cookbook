## Mat plot lib

```python
import pandas as pd
import numpy as np
```

### Common init:

```python
import matplotlib.pyplot as plt
```

```python
plt.close('all')
```

### Simple plot over values:

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts = ts.cumsum() # cumulative sum 
```

```python
ts.plot()
```

### Plot multiple columns of pandas dataframe:

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=list('ABCD'))
```

```python
df = df.cumsum()
```

```python
plt.figure();
```

```python
df.plot();
```

### plot one column versus another using the x and y keywords in plot():

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
```

```python
df3['A'] = pd.Series(list(range(len(df3)))) # lenght of the dataframe df3
```

```python
df3.plot(x='A', y='C')
```
